```python
import pandas as pd

if __name__ == '__main__':
    target = 'Model'
    tables = ['table' + target + 'GNB.xlsx', 'table' + target + 'DecisionTree.xlsx', 
              'table'+ target + 'RandomForest.xlsx', 'table' + target + 'KNN.xlsx' ]
#     tables = ['table' + target + 'DecisionTree.xlsx']
    table = pd.read_excel(tables[0])
    
    devices = table[target].unique()
    for tableName in tables:
        table = pd.read_excel(tableName)
        smallerTable = pd.DataFrame(columns = table.columns)
        i = 0
        for device in devices:
            smallerTable = smallerTable.append(table[table[target] == device].mean(), ignore_index=True)
            smallerTable['Std Conf'][i] = table[table[target] == device]['Avg Conf'].std()
            i+=1
        smallerTable[target] = devices
       
        smallerTable = smallerTable.dropna()
        smallerTable = smallerTable.reset_index(drop = True)
        print(smallerTable)
        smallerTable.to_excel("smaller" + tableName)
    
```
